```python
df['Diff_Eomth01'] = df['Totpaid Eomth01'] - df['Due Eomth01']
df['Diff_Eomth02'] = df['Totpaid Eomth02'] - df['Due Eomth02']
df['Diff_Eomth03'] = df['Totpaid Eomth03'] - df['Due Eomth03']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=df[['Diff_Eomth01', 'Diff_Eomth02', 'Diff_Eomth03', 'Nps Score']])
plt.title("Payment Differences Over 3 Months vs NPS Score")
plt.xlabel("Data Points")
plt.ylabel("Amount Differences (Paid - Due) / NPS Score")
plt.legend(["Diff_Eomth01", "Diff_Eomth02", "Diff_Eomth03", "Nps Score"])
plt.grid(True)
plt.show()
```

```python
# Plot 2: NPS Score Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Nps Score'], bins=10, kde=True)
plt.title("NPS Score Distribution")
plt.xlabel("NPS Score")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
